```python
## CUstom MOdule
```

```python
class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.hidden = nn.LazyLinear(256)
        self.out = nn.LazyLinear(10)
    def forward(self,X):
        return self.out(F.relu(self.hidden(X)))
```

```python
net = MLP()
net(X).shape
```

```python
## The Sequential Module
```

```python
class MySequential(nn.Module):
    def __init__(self,*args):
        super().__init__()
        for idx,module in enumerate(args):
            self.add_module(str(idx),module)
    def forward(self,X):
        for module in self.children():
            X = module(X)
        return X
```

```python
net = MySequential(nn.LazyLinear(256),nn.ReLU(),nn.LazyLinear(10))
```

```python
net(X).shape
```

```python
## Executing code in the forward propogation method
```

```python
class FixedHiddenMLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.rand_weight = torch.rand((20,20))
        self.linear = nn.LazyLinear(20)

    def forward(self,X):
        X = self.linear(X)
        X = F.relu(X @ self.rand_weight + 1)
        X = self.linear(X)

        #control flow
        while X.abs().sum() >1: #norm
            X /=2
        return X.sum()
```

```python
net = FixedHiddenMLP()
net(X)
```

```python
## Nest MLP
```

```python
class NestMLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(nn.LazyLinear(64),nn.ReLU(),
                                 nn.LazyLinear(32),nn.ReLU())
        self.linear = nn.LazyLinear(16)

    def forward(self,X):
        return self.linear(self.net(X))
```

```python
chimera = nn.Sequential(NestMLP(),nn.LazyLinear(20),FixedHiddenMLP())
chimera(X)
```

## ParallelModule

```python
import torch 
import torch.nn as nn
```

```python
class ParallelModule(nn.Module):
    def __init__(self,net1:nn.Module,net2:nn.Module,dim:int ):
        super(ParallelModule,self).__init__()
        self.net1 = net1
        self.net2 = net2
        self.dim = dim
    def forward(self,x):
        out1 = self.net1(x)
        out2 = self.net2(x)
        return torch.cat([out1,out2],dim=self.dim)
        
        
```

```python
net1 = nn.Sequential(
    nn.Conv2d(3, 16, kernel_size=3, padding=1),
    nn.ReLU()
)

net2 = nn.Sequential(
    nn.Conv2d(3, 32, kernel_size=5, padding=2),
    nn.ReLU()
)

parallel = ParallelModule(net1, net2, dim=1)

x = torch.randn(8, 3, 64, 64)  # batch of 8 RGB images
y = parallel(x)
```
